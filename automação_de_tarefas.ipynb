```python
import yfinance as yf
import pyautogui
import pyperclip
```

#### Passo 1 - buscas dados das ações automaticamente

```python
ticker = input("Digite o código da ação: ")
dados = yf.Ticker(ticker).history("6mo")
fechamento = dados.Close
fechamento.plot()
```

#### Passo 2 - realizar as análises solicitadas pelo meu gestor

- Cotação máxima
- Cotação mínima
- Cotação atual

```python
maxima = round(fechamento.max(), 2)
minima = round(fechamento.min(), 2)
atual = round(fechamento[-1], 2)

print(maxima)
print(minima)
print(atual)    
```

#### Passo 3: enviar automaticamente o e-mail com o resultado da análise

- Abrir uma nova aba (ctrl + T)
- Digitar www.gmail.com e dar enter
- Clicar no botão Escrever
- Digitar destinatário (tab)
- Digitar o assunto (tab)
- Preencher o corpo do e-mail
- Clicar no botão enviar

#### Criando o e-mail à ser enviado

```python
destinatario = "xxxxx@gmail.com"
assunto = "Análise diária"

mensagem = f"""
Bom dia,

Segue abaixo as análises da ação {ticker} dos últimos seis meses:

Cotação máxima: R${round(maxima,2)}
Cotação mínima: R${round(minima,2)}
Cotação atual: R${round(atual,2)}

Atenciosamente,
Seu nome.
"""
```

```python
print(destinatario)
print(assunto)
print(mensagem)
```

#### Automatizando o envio

```python
# dar uma pausa de 5 segundos entre as ações do pyautogui
pyautogui.PAUSE = 5

# Abrir uma nova aba (ctrl + T)
pyautogui.hotkey("ctrl", "t")

# Copiar o endereço da caixa de e-mail e dar enter (inseir o seu endereço)
pyperclip.copy("www.gmail.com")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("enter")

# Clicar no botão Escrever
pyautogui.click(x=117, y=206)

# Digitar destinatário (tab)
pyperclip.copy(destinatario)
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("tab")

# Preencher o assunto (tab)
pyperclip.copy(assunto)
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("tab")

# Preencher o corpo do e-mail
pyperclip.copy(mensagem)
pyautogui.hotkey("ctrl", "v")

# Clicar no botão Enviar
pyautogui.click(x=1136, y=979)

# fechar a aba do gmail
pyautogui.hotkey("ctrl", "f4")

# Imprimir mensagem de enviado com sucesso
print('E-mail enviado com sucesso!')
```
